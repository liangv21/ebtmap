```python
import geopy.distance
import pandas as pd
import numpy as np
```

- source for co2 emissions https://www.carbonindependent.org/22.html#:~:text=CO2%20emissions%20from%20aviation%20fuel,CO2%20per%20passenger%20per%20hour.

- source for carbon footprint per person per year (4 tons = 3628.74 kg)
https://www.nature.org/en-us/get-involved/how-to-help/carbon-footprint-calculator/#:~:text=A%20carbon%20footprint%20is%20the,is%20closer%20to%204%20tons.

```python
flight_df = pd.read_csv('output.csv')
```

```python
flight_df.head()
```

```python
bos = (42.3656, -71.0096)
jfk = (40.6413, -73.7781)
coord_df = flight_df[['lat', 'lon']]
flight_df['dist'] = np.nan
flight_df['co2KGPerPerson'] = np.nan
# percentage of co2 emission of person in this flight taking in kg per person per year
flight_df['percentageco2(%)'] = np.nan
for i in range(160):
    coords_2 = (coord_df.iloc[i, 0], coord_df.iloc[i, 1])
    dist = round(geopy.distance.geodesic(bos, coords_2).km, 2)
    flight_df['dist'].iloc[i] = dist
    co2 = round(((dist * 115)/1000), 2)
    flight_df['co2KGPerPerson'].iloc[i] = co2
    flight_df['percentageco2(%)'].iloc[i] = co2/3628.74*100
    

for i in range(160, len(flight_df)):
    coords_2 = (coord_df.iloc[i, 0], coord_df.iloc[i, 1])
    dist = round(geopy.distance.geodesic(jfk, coords_2).km, 2)
    flight_df['dist'].iloc[i] = dist
    co2 = round(((dist * 115)/1000), 2)
    flight_df['co2KGPerPerson'].iloc[i] = co2
    flight_df['percentageco2(%)'].iloc[i] = co2/3628.74*100   
     
```
